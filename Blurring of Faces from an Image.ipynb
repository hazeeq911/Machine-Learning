```python
#A path is created to read image files
path = "/contentgdrive/MyDrive/images/"
```

```python
#Import libraries
import cv2
import numpy as np
#From PIL import Image (Python Imaging Library (PIL), which adds support for opening, manipulating, and saving images)
import matplotlib.pyplot as plt 
from PIL import Image 
%matplotlib inline
from IPython.display import Image
```

```python
#A face cascade which is an Object Detection Algorithm used to identify faces in an image or a real time video is applied
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

#Images are read
img1 = cv2.imread(path+"first expression.jpg")
img2 = cv2.imread(path+"second expression.jpg")
img3 = cv2.imread(path+"third expression.jpg")

#Images are converted from BGR to RGB
img_rgb1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img_rgb2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img_rgb3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

#Detection functions are made
detections1 = face_cascade.detectMultiScale(img1,scaleFactor=1.1,minNeighbors=6)
detections2 = face_cascade.detectMultiScale(img2,scaleFactor=1.1,minNeighbors=6)
detections3 = face_cascade.detectMultiScale(img3,scaleFactor=1.1,minNeighbors=6)

#The width and height of the face recognized can be checked by using the print function as of below
print(detections1)
print(detections2)
print(detections3)
```

```python
#Face blur is applied by integrating the face_cascade function for the first image
for face in detections1:
  x,y,w,h = face

  #Array is sliced so that recognised face is blurred
  img_rgb1[y:y+h,x:x+w] = cv2.GaussianBlur(img_rgb1[y:y+h,x:x+w],(15,15),cv2.BORDER_DEFAULT)

  cv2.rectangle(img_rgb1,(x,y),(x+w,y+h),(255,0,0),2)

import matplotlib.image as mpimg 
from matplotlib.pyplot import imshow
%matplotlib inline

imshow(img_rgb1)

cv2.waitKey(0)
```

```python
#Face blur is applied by integrating the face_cascade function for the the second image
for face in detections2:
  x,y,w,h = face

  #Array is sliced so that recognised face is blurred
  img_rgb2[y:y+h,x:x+w] = cv2.GaussianBlur(img_rgb2[y:y+h,x:x+w],(15,15),cv2.BORDER_DEFAULT)

  cv2.rectangle(img_rgb2,(x,y),(x+w,y+h),(255,0,0),2)

import matplotlib.image as mpimg 
from matplotlib.pyplot import imshow
%matplotlib inline

imshow(img_rgb2)

cv2.waitKey(0)
```

```python
#Face blur is applied by integrating the face_cascade function for the the third image
for face in detections3:
  x,y,w,h = face

  #Array is sliced so that recognised face is blurred
  img_rgb3[y:y+h,x:x+w] = cv2.GaussianBlur(img_rgb3[y:y+h,x:x+w],(15,15),cv2.BORDER_DEFAULT)

  cv2.rectangle(img_rgb3,(x,y),(x+w,y+h),(255,0,0),2)

import matplotlib.image as mpimg 
from matplotlib.pyplot import imshow
%matplotlib inline

imshow(img_rgb3)

cv2.waitKey(0)
```
